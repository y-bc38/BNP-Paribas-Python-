```python
def get_date(year):
    
    year = str(year)
    
    response = requests.get("https://www.federalreserve.gov/newsevents/pressreleases/" + year + "-press-fomc.htm")
    soup = bs4.BeautifulSoup(response.text, "html.parser")
    
    date = soup.find_all("time")

    date_list = []
    date_list2 = []         #list2是去掉重複日期的
    date_list3 = []         #list3為轉換格式後

    for i in date:
        date_list.append(i.text)
        
        for i in date_list:
            if i not in date_list2:
                date_list2.append(i)
                
    for i in date_list2 : 
        day = datetime.strptime(i, '%m/%d/%Y')
        new_day =  day.strftime("%Y%m%d")
        date_list3.append(new_day)
    
    return date_list3
```

```python
date_list_2019 = []
date_list_2018 = []
date_list_2017 = []

for i in [20191211, 20191031, 20190918, 20190731, 20190619, 20190501, 20190320, 20190130]:
    date_list_2019.append("".join(str(i)))

for i in [20181219, 20181108, 20180926, 20180801, 20180613, 20180502, 20180321, 20180131]:
    date_list_2018.append("".join(str(i)))
    
for i in [20171213, 20171101, 20170920, 20170726, 20170614, 20170503, 20170315, 20170201]:
    date_list_2017.append("".join(str(i)))
```

```python
date_list_2022 = get_date(2022)
date_list_2021 = get_date(2021)
date_list_2020 = get_date(2020)

date_list_url = date_list_2022 + date_list_2021 + date_list_2020 + date_list_2019 + date_list_2018 + date_list_2017
date_list_url
```

```python
# 抓取近2年會議紀錄
for i in range(len(url_list)):
    response = requests.get(url_list[i])
    soup = bs4.BeautifulSoup(response.text, "html.parser")
    
    #用find_all會產生list，用list囊括所有資料
    content = soup.find(class_ = "col-xs-12 col-sm-8 col-md-8")
    content = content.text

    #文中的亂碼(可去找全部的列表)
    content = content.replace("â\x80\x91", "-")
    content = content.replace("â\x80\x99", "'")

    content = content.replace("U.S.", "US")
    content = content.replace("\xa0", " ")

    #照句子分且刪除會議人員
    content = content.split(".")

    content = "\n".join(content)
    content

    text_create(date_list_url[i], content)
```
